```python
import pandas as pd
import numpy as np

# Load the data
data = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/DADS5001-Tools/Assignment3_DMdataset.csv")

# Check information of dataset
print(data.info())

# Calculate the average for each column
avg_values = data.mean()

# Replace the zero values with the average
data = data.replace(0, avg_values)

# Convert categorical variables to numerical
data['Glucose'] = data['Glucose'].astype(float)
data['BloodPressure'] = data['BloodPressure'].astype(float)
data['SkinThickness'] = data['SkinThickness'].astype(float)
data['Insulin'] = data['Insulin'].astype(float)
data['BMI'] = data['BMI'].astype(float)
data['DiabetesPedigreeFunction'] = data['DiabetesPedigreeFunction'].astype(float)
data['Age'] = data['Age'].astype(float)

# Scale the variables
data = (data - data.min()) / (data.max() - data.min())

# Show data table
data.head()
```

```python
import plotly.express as px

# Calculate the correlation matrix
corr_matrix = data.corr()

# Create the heatmap
fig = px.imshow(corr_matrix, color_continuous_scale='Blues')

# Add labels
fig.update_layout(
    title='Correlation Heatmap for Diabetes Dataset',
    xaxis=dict(tickvals=list(range(len(corr_matrix.columns))),
               ticktext=corr_matrix.columns,
               title='Features'),
    yaxis=dict(tickvals=list(range(len(corr_matrix.columns))),
               ticktext=corr_matrix.columns,
               title='Features'),
    height=600,
    width=600
)

fig.show()
```

Based on the heatmap correlation, we can observe that some variables have a strong positive correlation with the target variable (Outcome), such as Glucose and BMI. This suggests that higher values of these variables are associated with a higher likelihood of having diabetes. On the other hand, some variables have a weak or no correlation with the target variable, such as BloodPressure,  Insulin and SkinThickness.

It is important to note that correlation does not imply causation, and further analysis is needed to determine the causal relationships between variables. However, the heatmap correlation provides a useful starting point for exploring the relationships between variables in the dataset.
